```python
#combine review text and summary into one column and make everything lowwer case 

#Does combination become neutral?

Sports_reviews["textonly"] = Sports_reviews["reviewText"].map(str) + Sports_reviews["summary"]
Sports_reviews.apply(lambda x: x.astype(str).str.lower())
```

```python
#Positive reviews

positive = ["exiciting", "good", "recommended", "love", "awesome", "best", "excellent", "highly", "amazing", "great",
            "fun", "recommended", "nice", "impressive", "impressed", "useful", "fantastic", "super", "superb", "unreal", 
           "top-quality", "top-notch", "thrilling", "terrific", "supreme", "stylish", "splendid", "remarkable"]

for key in positive:
    Sports_reviews[key] = (Sports_reviews.textonly.str.contains(key, case=False)).astype(int)
Sports_reviews
```

```python
target = Sports_reviews['positive'] = np.where((Sports_reviews['overall'] >=3.7),1,0)
```

```python
Sports_reviews.head(5)
```

```python
# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

data = Sports_reviews[positive]

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))

#Confusion matrix!

print((1-(target != y_pred).sum()/data.shape[0])*100)
```

```python
y_train = Sports_reviews['positive']
```

```python
X_train = Sports_reviews[["exiciting", "good", "recommended", "love", "awesome", "best", "excellent", "highly", "amazing", "great",
            "fun", "recommended", "nice", "impressive", "impressed", "useful", "fantastic", "super", "superb", "unreal", 
           "top-quality", "top-notch", "thrilling", "terrific", "supreme", "stylish", "splendid", "remarkable"]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, random_state=42, test_size=0.40)
```

```python
from sklearn.metrics import accuracy_score

rfc = ensemble.RandomForestClassifier(random_state=42)
rfc.fit(X_train,y_train)
cv = cross_val_score(rfc, X_train, y_train, cv=5)



#Test
y_true, y_pred = y_test, rfc.predict(X_test)
print("Test Set score = {:.3} ".format(accuracy_score(y_true, y_pred)))


print("Training set Cross validation = {}".format(cv))
print("mean = {:.3}".format(cv.mean()))
```
