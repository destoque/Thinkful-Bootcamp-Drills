```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('C:\\Users\\david\\Downloads\\listings.csv')
df
```

```python

plt.figure(figsize=(15, 8))
sns.violinplot(x=df['price'], color='green', alpha=.2)
plt.title('Average Airbnb Rental price in San Diego')
plt.xlabel('Price in USD')


plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.set(style="whitegrid")

ax = sns.boxplot(x=df['price'])  

plt.title('Average Airbnb Rental price in San Diego')

sns.despine(offset=10, trim=True)
ax.set(xlabel='Price in USD', ylabel='')

plt.show()
```

```python
plt.figure(figsize=(20, 10))

ax = sns.boxplot(x='room_type',y='price',data=df)  
plt.title('Price by Rental Type')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Price')

plt.show()


print ('Entire home rentals make the graph looking funny as they have a large amount of variabilty in price')
```

```python
sns.set(style="darkgrid")

g = sns.factorplot(x="room_type", y="price", data=df,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Price")
g.set_xlabels("")
plt.title('Average Price per Rental Type')
plt.show()
```
